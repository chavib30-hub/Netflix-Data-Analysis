```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#reading Netflix csv file

df=pd.read_csv("netflix_titles.csv")
df
```

```python
#checking null values
df.isnull().sum()
```

```python
#replacing null vallues 
df["director"]=df["director"].fillna("Unknown")
df["cast"]=df["cast"].fillna("Not Available")
df["country"]=df["country"].fillna("Unknown")
df["rating"]=df["rating"].fillna("Not Rated")
df=df.dropna(subset=["date_added"])
df=df.dropna(subset=["duration"])

df
```

```python
#now again checking null values 
df.isnull().sum()
```

# Basic Business Questions

```python
#Count of Movies and  Tv shows

df.groupby("type").size()


```

```python
#which type is more
df.groupby("type").size().idxmax()
```

```python
#which country produces most content

def c_content(row):
    if row["type"]=='Movie':
        return 1
    else:
        return int (row["duration"].split(" ")[0])

df["content"]=df.apply(c_content , axis=1)   

df.groupby("country")["content"].size().idxmax()
```

```python
#top 10 content producing countries
def c_content(row):
    if row["type"]=='Movie':
        return 1
    else:
        return int (row["duration"].split(" ")[0])

df["content"]=df.apply(c_content , axis=1)   

df1=df.groupby("country")["content"].size().sort_values(ascending=False)
top10=df1.head(10).reset_index()

sns.barplot(data=top10,x="country",y="content")
plt.xticks(rotation=90)
plt.xlabel("Country")
plt.ylabel("Content")
plt.title(" Top 10 most Content producing countries")
plt.show()
```

```python
#most common rating

df.groupby("rating").size().idxmax()
```

# Trend based  questions

```python
#netflix content grown over years

y_count=df.groupby("release_year")["title"].count()

plt.plot(y_count.index,y_count.values)
plt.xlabel("Release years")
plt.ylabel("Titles")
plt.title("Netflix content grown over years")
plt.show()
```

```python
#highest number of releases


def c_content(row):
    if row["type"]=='Movie':
        return 1
    else:
        return int (row["duration"].split(" ")[0])

df["content"]=df.apply(c_content , axis=1) 

year=df.groupby("release_year")["content"].size().idxmax()
print(int(year))
```

```python
#when add most content
df["date_added"]=pd.to_datetime(df["date_added"].iloc[0])
df["year"]=df["date_added"].dt.year
year=df.groupby("year")["content"].size().idxmax()
print(int(year))
```

```python
#Are movies increasing?

y_count=df[df["type"]=="Movie"].groupby("release_year").size()
x_count=df[df["type"]=="TV Show"].groupby("release_year").size()
plt.plot(y_count.index,y_count.values)
plt.plot(x_count.index,x_count.values)
plt.xlabel("release_year")
plt.ylabel("type")
plt.legend(["Movie","TV Show"])
plt.title("Movies Vs TV Show")
plt.show()
```
